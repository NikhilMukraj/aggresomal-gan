```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2

# https://stackabuse.com/k-means-clustering-with-scikit-learn/
```

```python
path = os.getcwd() + '\\augmented_data\\'

img_size = (64, 64)
trainX = np.zeros((72_800, img_size[0], img_size[1]))
```

```python
update_per = 10000

for n, i in enumerate(os.listdir(path)):
    filename = os.fsdecode(i)
    temp_img = cv2.imread(path + filename, 0) / 255
    temp_img = cv2.resize(temp_img, img_size)
    trainX[n] = temp_img
    if n % update_per == 0:
        print(f'Image #: {n}')
```

```python
x, y, z = trainX.shape
print('shape: ', trainX.shape)

trainX = trainX.reshape(x*y, z)
print('reshape: ', trainX.shape)
```

```python
total_clusters = 5

kmeans = MiniBatchKMeans(n_clusters = total_clusters)

kmeans.fit(trainX)
```

```python
Z = kmeans.predict(trainX)
```

```python
trainX = trainX.reshape((72800, 64, 64))

for n, i in enumerate(trainX):
    plt.imshow(i)
    plt.title(kmeans.labels_[n])
    plt.show()
```
